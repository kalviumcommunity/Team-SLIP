# 03 — Train-Test Split
Stratified 80/20 split preserving class distribution.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from src.config import (
    DATA_PROCESSED_PATH, NUMERICAL_FEATURES,
    CATEGORICAL_FEATURES, TARGET, RANDOM_STATE, TEST_SIZE
)
```

## 1. Load Clean Data and Separate Features/Target

```python
df = pd.read_csv(DATA_PROCESSED_PATH)
X = df[NUMERICAL_FEATURES + CATEGORICAL_FEATURES]
y = df[TARGET]
print(f"X shape: {X.shape}, y shape: {y.shape}")
print(f"Default rate: {y.mean()*100:.1f}%")
```

## 2. Stratified Split
`stratify=y` ensures both train and test sets have the same ~88/12 class ratio.
Without this, random chance could give the test set a different ratio,
making evaluation unreliable.

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)

print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
print(f"y_train: {y_train.shape}, y_test: {y_test.shape}")
print(f"\nTrain class distribution:\n{y_train.value_counts(normalize=True).round(4)}")
print(f"\nTest class distribution:\n{y_test.value_counts(normalize=True).round(4)}")
```

## 3. Verify Stratification

```python
assert abs(y_train.mean() - y_test.mean()) < 0.01, "Stratification failed!"
print(f"Train default rate: {y_train.mean()*100:.2f}%")
print(f"Test default rate:  {y_test.mean()*100:.2f}%")
print("✓ Stratification verified — both sets have same class ratio")
```

## 4. Save Split Data

```python
import os
os.makedirs('data/processed', exist_ok=True)

X_train.to_csv('data/processed/X_train.csv', index=False)
X_test.to_csv('data/processed/X_test.csv', index=False)
y_train.to_csv('data/processed/y_train.csv', index=False)
y_test.to_csv('data/processed/y_test.csv', index=False)

print("Saved X_train, X_test, y_train, y_test to data/processed/")
print(f"Train: {X_train.shape[0]:,} rows | Test: {X_test.shape[0]:,} rows")
```
